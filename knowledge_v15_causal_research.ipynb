# 🧠 Knowledge Base: CausalCrisis V15 (Q1 Upgrade)
Đây là tài liệu tóm tắt các thay đổi kiến trúc quan trọng nhằm giải quyết 4 điểm nghẽn (Bottlenecks) trong nghiên cứu nhân quả cho thảm họa thảm họa (Disaster Decision Support).

---

## 🛠 B1: Deep Residual Disentangler & HSIC
**Vấn đề:** Disentangler cũ quá nông (shallow), không đủ sức tách biệt đặc trưng nhân quả ($x_c$) và nhiễu ($x_s$).
**Giải pháp:**
- **Kiến trúc:** Sử dụng **3-block Residual Network** giúp tăng dung lượng mô hình.
- **HSIC (Hilbert-Schmidt Independence Criterion):** Thay thế loss trực giao tuyến tính bằng đo tính độc lập phi tuyến. Ép $x_c$ và $x_s$ không có bất kỳ mối quan hệ thống kê phức tạp nào.
- **Reconstruction Loss:** Đảm bảo $x_c$ và $x_s$ cùng nhau chứa đầy đủ thông tin của đặc trưng gốc.

## 🕸 B2: Heterophily-Aware GAT (HA-GAT)
**Vấn đề:** Đồ thị gốc có nhiều cạnh nối giữa các bài đăng khác lớp (Heterophily), gây nhiễu cho việc lan truyền thông tin.
**Giải pháp:**
- **Dynamic Causal Graph:** Xây dựng kNN dựa trên $x_c$ (đặc trưng sạch) thay vì raw features.
- **Pseudo-label Gating:** Sử dụng dự đoán nhãn tạm thời để dập tắt các cạnh nối giữa các node có xác suất thuộc lớp khác nhau.

## 🤝 B3: Cross-Modal Causal Alignment
**Vấn đề:** Sự mâu thuẫn giữa thông tin hình ảnh và văn bản (ví dụ: ảnh thảm họa nhưng text là quảng cáo).
**Giải pháp:**
- **Modal Causal Consistent Loss (MCL):** Áp dụng MSE loss giữa $x_{c,v}$ (hình ảnh) và $x_{c,t}$ (văn bản). Ép mô hình chỉ tin vào những bằng chứng nhân quả xuất hiện đồng thời ở cả hai phương thức.

## 🎲 B4: Monte Carlo Backdoor Adjustment (BA)
**Vấn đề:** Dataset bias cố hữu (ví dụ: lũ lụt thường đi kèm với cây cối) khiến mô hình học sai quy luật nhân quả.
**Giải pháp:**
- **Inference-time do-calculus:** Thực hiện $P(Y | do(X))$.
- **Efficient Sampling:** Lấy mẫu 100 confounder prototypes từ dataset để tính toán giá trị kỳ vọng (expectation) của dự đoán, giúp loại bỏ ảnh hưởng của bias môi trường.
